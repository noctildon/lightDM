```python
import numpy as np
import pandas as pd
```

## Read the files from Baishan

```python
# the 1st run (fewer states)
input_file = './eval_GamowTeller_Wavf-Xe131_Wavi-Cs131_jj55pna_hw7.8912_ph3,4,5_0_0-6,7_14_14.dat'
output_file = './Xe131_Cs131.dat'

# the 2nd run (more states)
input_file = 'eval_GamowTeller_Wavf-Xe131_Wavi-Cs131_jj55pna_hw7.8912_N100.dat'
output_file = 'Xe131_Cs131_morestates.dat'

cc_dat = pd.read_csv(input_file, delimiter=',', skipinitialspace=True)
cc_dat.info()
```

## Convert to excitation energy and GT strength

```python
cc_dat['Ex'] = cc_dat['Energy ket (MeV)'] - cc_dat['Energy bra (MeV)'] # in MeV
cc_dat['GT'] = cc_dat['Op exp val 1b']**2 / np.sqrt(2) # sqrt(2) is the convention
cc_dat = cc_dat[['Ex', 'GT']]
```

```python
cc_dat.head()
```

## Save to numpy file

```python
np.savetxt(output_file, cc_dat.to_numpy(), delimiter=' ', header='Excitation energy [MeV], GT strength')
```
